```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Making six differently distributed data sets
#distributionvar = np.random.distributionname([arguments], 100)
#document to lookup supported distributions
#https://docs.scipy.org/doc/numpy-1.14.0/reference/routines.random.html
normal = np.random.normal(0, 1, 100) #mean=0, std=1
poisson = np.random.poisson(1, 100) #lam=1
power = np.random.power(2,100) #(a[, size])
negative_binomial =np.random.negative_binomial(2,0.5,100) #(n, p[, size]), p<1!!
logistic = np.random.logistic(2,5,100)#([loc, scale, size])
geometric = np.random.geometric(0.5,100)#(p[, size])
```

```python
def show_Summary_stats(distributionData):
    #print(distributionData.__name__)
    print("Mean=", np.mean(distributionData))
    print("Std=", np.std(distributionData))
    #Make a histogram with a solid line at the mean and dashed lines for the std
    plt.hist(distributionData,color='c', bins=20)
    # Add a vertical line at the mean.
    plt.axvline(distributionData.mean(), color='b', linestyle='solid', linewidth=2)
    # Add a vertical line at one standard deviation above the mean.
    plt.axvline(distributionData.mean() + distributionData.std(), color='b', linestyle='dashed', linewidth=2)
    # Add a vertical line at one standard deviation below the mean.
    plt.axvline(distributionData.mean()-distributionData.std(), color='b', linestyle='dashed', linewidth=2) 
    plt.show()
```

```python
show_Summary_stats(normal)
```

```python
show_Summary_stats(poisson)
```

```python
show_Summary_stats(power)
```

```python
show_Summary_stats(negative_binomial)
```

```python
show_Summary_stats(logistic)
```

```python
show_Summary_stats(geometric)
```

Answer: Only for the normal and the logistic distribution the mean really describes the central tendency of the data. The other distributions lead to datasets, that do not cluster symmetrically.

Additionally:
Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.
Add them together to create a third variable.
Graph the third variable using a histogram.
Compute the mean and standard deviation and plot them as vertical lines on the histogram.
Evaluate the descriptive statistics against the data.

```python
data1 = np.random.normal(5, 0.5, 100)
data2 = np.random.normal(10,1,100)
data3 = data1 + data2

show_Summary_stats(data1)
show_Summary_stats(data2)
show_Summary_stats(data3)
```

Both datasets are normally distributed, but because the means of the datasets are different, the resulting data3 will not be normally distributed, meaning it will show some skew.
