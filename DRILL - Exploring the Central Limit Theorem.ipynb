```python
import numpy as np
import pandas as pd
import scipy
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 

# Letâ€™s make histograms for the two groups. 

plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()
```

1. Increase the size of your samples from 100 to 1000, then calculate the means and standard deviations for your new samples and create histograms for each. Repeat this again, decreasing the size of your samples to 20. What values change, and what remain the same?

Expectation: 
When incraesing from 100 to 1000, the mean will stay very similar but the std should decrease.
When decreasing to 20 samples, the mean will become less accurate, change a littlebit more than before, but the std should increase.

Result after computing
The mean indeed only changes a little, bit, but the standard deviation changed more but unexpectly. For population1 the std increased with more samples, but decrased with fewer samples. For poplation2 the std decreased in both cases.


```python
def show_sumStatsHist(sample1, sample2):
    print(sample1.mean())
    print(sample2.mean())
    print(sample1.std())
    print(sample2.std())
    
    print(ttest_ind(sample2, sample1, equal_var=False))

    plt.hist(sample1, alpha=0.5, label='Sample 1') 
    plt.hist(sample2, alpha=0.5, label='Sample 2') 
    plt.legend(loc='upper right') 
    plt.show()
```

```python
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

show_sumStatsHist(sample1,sample2)
```

```python
sample1 = np.random.choice(pop1, 1000, replace=True)
sample2 = np.random.choice(pop2, 1000, replace=True)

show_sumStatsHist(sample1,sample2)
```

```python
sample1 = np.random.choice(pop1, 20, replace=True)
sample2 = np.random.choice(pop2, 20, replace=True)

show_sumStatsHist(sample1,sample2)
```

2. Change the probability value (p in the NumPy documentation) for pop1 to 0.3, then take new samples and compute the t-statistic and p-value. Then change the probability value p for group 1 to 0.4, and do it again. What changes, and why?

Expectation: 
If the distributions of the populations become more similar, the p-value of the t-test, will become bigger, indicating that the samples are more likely to have been drawn from the same population.

Result:
The more similar the two populations, the bigger the p-value.

```python
pop1 = np.random.binomial(10, 0.3, 10000)
pop2 = np.random.binomial(10, 0.5, 10000)

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

show_sumStatsHist(sample1,sample2)
```

```python
pop1 = np.random.binomial(10, 0.4, 10000)
pop2 = np.random.binomial(10, 0.5, 10000)

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

show_sumStatsHist(sample1,sample2)
```
